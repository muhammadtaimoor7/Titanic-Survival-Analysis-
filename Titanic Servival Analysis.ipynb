```python
# TITANIC SURVIVAL ANALYSIS PROJECT
# PURPOSE:
# To analyze the Titanic dataset and find out
# which factors affected the survival of passengers.
# such as Gender, Age, Passenger Class and Fare.
# DATASET: tested.csv
# ROWS: 418
# COLUMNS: 12
# TOOLS USED:
# - Python 
# - Pandas
# - Matplotlib
```

```python
#STEP 1
#IMPORT BOTH LIBRARIES
import pandas as pd
import matplotlib.pyplot as plt
#LOAD CSV FILE 
df=pd.read_csv("tested.csv")
df
```

```python
#STEP 2
#CHECK FIRST FEW ROWS
df.head(4)
```

```python
#STEP 3
#CHECK LAST FEW ROWS
df.tail(6)
```

```python
#STEP 4
#TOTAL ROWS AND COLUMNS
df.shape
```

```python
#STEP 5
#DISPLAY ALL COLUMNS
df.columns
```

```python
#STEP 6
#CHECK DATA TYPES
df.dtypes
```

```python
#STEP 7
#COMPLETE DATASET INFORMATION
df.info()
```

```python
#STEP 8
#STATISTICAL SUMMARY
df.describe()
```

```python
#STEP 9
#CHECKING MISSING VALUES
df.isnull().sum()
```

```python
#STEP 10
#TOTAL MISSING VALUES
df.isnull().sum().sum()
```

```python
#STEP 11
#HIGHEST MISSING VALUEs IN ORDER
df.isnull().sum().sort_values(ascending=False)
#it will show the missing values in order
```

```python
#STEP 12
#HANDLE MISSING VALUES
#FIRST AGE COLUMN REPLACE MISSING WITH AVERAGE VALUE
df.fillna({"Age":df["Age"].mean()},inplace=True)
#FOR VERIFICATION
df['Age'].isnull().sum()
#OUTPUT SHOULD BE ZERO
#NOW
#FARE COLUMN
df.fillna({"Fare":df["Fare"].mean()},inplace=True)
df["Fare"].isnull().sum()
#NOW
#CABIN COLUMN HAS 78% MISSING VALUES
#SO WE WILL DROP CABIN
#USE INPLACE FOR REMOVING COL PERMANENTLY
df.drop(columns=["Cabin"],inplace=True)
#SO IT WILL DROP CABIN COLUMN  
```

```python
#STEP 13
# CHECK DUPLICATES
df[df.duplicated()]
```

```python
#STEP 14
#REMOVE DUPLICATED
df.drop_duplicates(inplace=True)
```

```python
#STEP 15
#VERIFY SHAPE AFTER DOING ALL THESE
df.shape
#11 here bcz we have removed cabin column
```

```python
#STEP 16
#ANALYZE SURVIVAL COLUMN 
df["Survived"]
```

```python
#STEP 17
#COUNT SERVIVERS
df["Survived"].value_counts()
```

```python
#STEP 18
#CREATE PIE CHART OF SURVIVAL RATE
#PIE SHOWS RATIO OR PERCENTAGE AND CONTAINS NEITHER X NOR Y AXIS
survival = df["Survived"].value_counts()
plt.pie(survival.values, labels=["Non Survived", "Survived"])
plt.show()
```

```python
#STEP 19
#ANALYZE SEX COLUMN
df["Sex"]
```

```python
#STEP 20
#COUNT MALE AND FEMALE
df["Sex"].value_counts()
```

```python
#STEP 21
#CREATE A BAR CHAT
#GENDER VS SURVIVAL
#GROUPBY USED FOR BOTH COLUMNS
gender=df.groupby('Sex')['Survived'].value_counts()
gender.plot(kind='bar')
plt.title("Gender vs Survival")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
#STEP 22
#ANALYZE CLASS COLUMN 
df["Pclass"]
```

```python
#STEP 23 
#COUNT VALUES
df["Pclass"].value_counts()
```

```python
#STEP 24
#BAR CHAT BW CLASS AND SURVIVAL
check=df.groupby('Pclass')['Survived'].value_counts()
check.plot(kind="bar")
plt.title("Class vs Survived")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
#0 in bar will show not survived
#1 will show survived
```

```python
#STEP 25
#ANALYZE AGE COL
df["Age"]
```

```python
#STEP 26
#HISTOGRAM AGE DISTRIBUTION
#NO VARIABLE NEEDED JUST STRAIGHT FORWAD PLOTTING
#MUST USE OF BINS IN HISTO
plt.hist(df["Age"],bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
#STEP 27
#INSIGHTS
# Insight 1: Majority passengers were between 20-35 years old
# Insight 2: Very few elderly passengers (60+) were on board
# Insight 3: Some children (0-10) were also present
```

```python
#STEP 28
#BOX PLOT
#AGE VS SURVIVAL
#BOX PLOT WORKS AS
#df.boxplot(column, by)
df.boxplot(column='Age', by='Survived')
plt.title("Age vs Survival")
plt.xlabel("Survived (0=No, 1=Yes)")
plt.ylabel("Age")
plt.show()
```

```python
#STEP 29
#ANALYZE FARE COLUMN
df["Fare"]
```

```python
#STEP 30
#SCATTER PLOT
#FARE VS AGE
#FOR SCATTER PLOT NO GROUPBY
#DIRECTLY 2 COLUMNS ARE LISTED
import matplotlib.pyplot as plt
plt.scatter(df['Age'], df['Fare'])
plt.title("Fare vs Age")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
#STEP 31
#SUBPLOT
#PLOT ALL THE GRAPHS ADDED IN THIS PROJECT IN A SINGLE IMAGE
#FIGSIZE INCLUDES(WIDTH,HEIGHT)
#IF GRAPH NUMBER ARE MORE THAN FOUR USE(3,2,1) STREAM AND SO ON
#LESS THEN FOUR USE(2,2,1) STREAM
#WRITE SUBPLOT() AT TOP OF EACH GRAPH AND SO ON TO OTHERS
#FIGSIZE INCLUDES(WIDTH,HEIGHT)
plt.figure(figsize=(14,12))

plt.subplot(3,2,1)
gender = df.groupby('Sex')['Survived'].value_counts()
gender.plot(kind='bar')
plt.title("Gender vs Survival")

plt.subplot(3,2,2)
check = df.groupby('Pclass')['Survived'].value_counts()
check.plot(kind="bar")
plt.title("Class vs Survived")

plt.subplot(3,2,3)
plt.hist(df["Age"], bins=20)
plt.title("Age Distribution")

ax4 = plt.subplot(3,2,4)
df.boxplot(column='Age', by='Survived', ax=ax4)
plt.title("Age vs Survival")
plt.suptitle("")

plt.subplot(3,2,5)
plt.scatter(df['Age'], df['Fare'])
plt.title("Fare vs Age")

plt.tight_layout() #for adjust spacing
plt.show()

```

```python
# STEP 32
# SUMMARY KEY FINDINGS
print("Titanic Survival Analysis Summary")
print(f"Total Passengers: {df.shape[0]}")
print(f"Survived: {df['Survived'].value_counts()[1]}")
print(f"Not Survived: {df['Survived'].value_counts()[0]}")
print(f"Average Age: {df['Age'].mean():.1f}") #1f gives 1 value only after decimal
print(f"Average Fare: {df['Fare'].mean():.1f}")
```

```python
#STEP 33
# OBSERVATIONS
# 1. Majority of passengers did not survive
# 2. Females had a higher survival rate than males
# 3. 1st class passengers survived more than 3rd class
# 4. Most passengers were between 20-35 years old
# 5. 3rd class had the highest number of deaths
# 6. Higher fare passengers had better survival chances
```
